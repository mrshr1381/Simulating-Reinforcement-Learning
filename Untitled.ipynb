```python
import numpy as np
import gymnasium as gym
import matplotlib.pyplot as plt

def train_q_learning(alpha, gamma, epsilon_decay, episodes=5000):
    env = gym.make("FrozenLake-v1", is_slippery=False)
    num_states = env.observation_space.n
    num_actions = env.action_space.n
    Q = np.zeros((num_states, num_actions))
    epsilon = 1.0
    min_epsilon = 0.01
    max_steps = 100

    for ep in range(episodes):
        state, _ = env.reset()
        for _ in range(max_steps):
            if np.random.uniform(0, 1) < epsilon:
                action = env.action_space.sample()
            else:
                action = np.argmax(Q[state])
            next_state, reward, terminated, truncated, _ = env.step(action)
            Q[state, action] += alpha * (reward + gamma * np.max(Q[next_state]) - Q[state, action])
            state = next_state
            if terminated or truncated:
                break
        epsilon = max(min_epsilon, epsilon * epsilon_decay)

    # Ø§Ø±Ø²ÛŒØ§Ø¨ÛŒ
    successes = 0
    test_episodes = 100
    for _ in range(test_episodes):
        state, _ = env.reset()
        for _ in range(max_steps):
            action = np.argmax(Q[state])
            next_state, reward, terminated, truncated, _ = env.step(action)
            state = next_state
            if terminated or truncated:
                if reward == 1:
                    successes += 1
                break

    return successes / test_episodes  # Ø¯Ø±ØµØ¯ Ù…ÙˆÙÙ‚ÛŒØª

# ---------------------
# Ù…Ù‚Ø§Ø¯ÛŒØ± Ù¾Ø§Ø±Ø§Ù…ØªØ±Ù‡Ø§
alpha_values = [0.1, 0.4, 0.8]
gamma_values = [0.8, 0.9, 0.99]
epsilon_decay_values = [0.99, 0.995, 0.999]

# ---------------------
# Ø«Ø¨Øª Ù†ØªØ§ÛŒØ¬
alpha_results = [train_q_learning(alpha=a, gamma=0.99, epsilon_decay=0.995) for a in alpha_values]
gamma_results = [train_q_learning(alpha=0.8, gamma=g, epsilon_decay=0.995) for g in gamma_values]
epsilon_decay_results = [train_q_learning(alpha=0.8, gamma=0.99, epsilon_decay=e) for e in epsilon_decay_values]

# ---------------------
# Ø±Ø³Ù… Ù†Ù…ÙˆØ¯Ø§Ø±Ù‡Ø§
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

axs[0].plot(alpha_values, alpha_results, marker='o')
axs[0].set_title("(alpha)")
axs[0].set_xlabel("alpha")
axs[0].set_ylabel("succession rate")
axs[0].grid(True)

axs[1].plot(gamma_values, gamma_results, marker='o', color='orange')
axs[1].set_title(" (gamma)")
axs[1].set_xlabel("gamma")
axs[1].set_ylabel("succession rate")
axs[1].grid(True)

axs[2].plot(epsilon_decay_values, epsilon_decay_results, marker='o', color='green')
axs[2].set_title("(epsilon_decay)")
axs[2].set_xlabel("epsilon_decay")
axs[2].set_ylabel("succession rate")
axs[2].grid(True)

plt.tight_layout()
plt.show()
```
